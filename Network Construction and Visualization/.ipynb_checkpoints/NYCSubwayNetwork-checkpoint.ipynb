```python
import networkx as nx 
import matplotlib.pyplot as plt 
import pandas as pd
import geopandas as gpd
import numpy as np
%matplotlib inline 
```

```python
stations = pd.read_csv('https://raw.githubusercontent.com/CUSP2021ADS/Data/main/NYCSubwayNodes.csv'); stations
```

```python
connections = pd.read_csv('https://raw.githubusercontent.com/CUSP2021ADS/Data/main/NYCSubwayEdges.csv'); connections
```

```python
destinations = stations.merge(connections, left_on = 'id', right_on = 'dest') \
.loc[:,['name', 'dest', 'lat', 'long', 'lines']] \
.drop_duplicates()

destinations.head()
```
